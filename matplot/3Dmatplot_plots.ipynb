```python
#3D matplotlib
# 3d plotting https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html
# https://rukshanpramoditha.medium.com/creating-interactive-3d-plots-in-matplotlib-7200e7514fec
# https://stackoverflow.com/questions/43490330/a-slider-doesnt-change-a-value-in-a-3d-plot

# for creating a responsive plot
%matplotlib widget
  
# importing required libraries
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import os
```

```python
def gaussian(x, y, sigma, mu):
    return (1 / (sigma * np.sqrt(2 * np.pi)) ) * np.power(np.e, -0.5 * ((x + y - mu)/sigma)**2)

fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-100, 100, 20)
y = np.linspace(-100, 100, 20)

def Plot3DGaussian(sigma, mu):

    X, Y = np.meshgrid(x, y, sparse=True)
    Z = gaussian(X, Y, sigma, mu)

    ax.clear()
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    fig.canvas.draw_idle()


def createSlider(sValue, sMin, sMax, sStep):
    slider = widgets.FloatSlider(value=sValue, min=sMin, max=sMax, step=sStep)
    return slider

interact(Plot3DGaussian, sigma=createSlider(1, 20, 200, 0.01), mu=createSlider(1, -100, 100, 0.01))
```
